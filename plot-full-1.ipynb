```python
import numpy as np
import matplotlib as plt
import matplotlib.pyplot as plt1
x = np.linspace(-5,5,100)
y = np.sin(x)
plt1.plot(x, y, label = 'sin(x)')
plt.pyplot.xlabel('x')
plt.pyplot.ylabel('y')
plt.pyplot.title('sin wave')
plt.pyplot.legend()
plt.pyplot.grid()
plt.pyplot.xlim(-5, 6)
plt.pyplot.ylim(-2,2)
plt.pyplot.show()
```
